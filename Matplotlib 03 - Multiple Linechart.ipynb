## We will be plotting multiple linecharts in a single graph
#### Suppose, we have to plot exam scores of two terms and average score as well.

```python
from matplotlib import pyplot as plt

# we can plot charts inline with following:
%matplotlib inline  
```

```python
first_term = [70, 80, 76, 89, 61, 60, 57]
second_term = [72,79,66, 88,65, 72, 81]
avg = [int((x+y)/2) for x,y in zip(first_term, second_term)]
avg
```

```python
# get the index for x axis
xs = [i for i, _ in enumerate(first_term)]
xs
```

```python
plt.plot(xs, first_term, 'g-', label='first') # green solid line
plt.plot(xs, second_term,  'r-.', label='second') # red dot-dashed line
plt.plot(xs, avg, 'b:', label='avg') # blue dotted line

plt.legend(loc=9)
plt.xlabel("exam marks")
plt.title("term marks variation")
plt.show()
```
