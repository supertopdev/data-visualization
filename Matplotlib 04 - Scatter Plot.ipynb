### Scatter plot helps you understand the relationship between two data sets.

#### In the following example, we want to see the relationship between students final exam score and the time span they spent in the library(in min)

```python
from matplotlib import pyplot as plt

# we can plot charts inline with following:
%matplotlib inline  
```

```python
marks = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
library_time = [175, 170, 205, 120, 220, 130, 105, 145, 190]
names = ['john', 'cisa', 'kapil', 'dane', 'eson', 'fany', 'gandu', 'hari', 'emly']
```

```python
plt.scatter(marks, library_time)

# label each point
for label, mark_count, time_count in zip(names, marks, library_time):
    plt.annotate(label, xy=(mark_count, time_count), 
        xytext=(5, -5),
        textcoords='offset points')
plt.title("Marks distribution vs.no of minutes spend in library")
```
